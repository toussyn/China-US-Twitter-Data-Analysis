```python
# Import packages
import json
import pandas as pd
import re

# String of path to file: tweets_data_path
tweets_data_path = '../input/global-twitter-data/global_twitter_data.json'

# Initialize empty list to store tweets: tweets_data
tweets_data = []

# Open connection to file
tweets_file = open(tweets_data_path, "r")

# Read in tweets and store in list: tweets_data
for line in tweets_file:
    tweet = json.loads(line)
    tweets_data.append(tweet)

# Close connection to file
tweets_file.close()

def extract_twitter_source(source):
        res = re.split('<|>', source)[2].strip()
        return res

#Convert to ddataframe
df1 = pd.DataFrame(tweets_data, columns =['created_at', 'full_text', 'source', 'user', 'retweet_count', 'lang'])
```
